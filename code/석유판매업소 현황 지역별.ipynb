```python
import requests
from bs4 import BeautifulSoup as bs
from urllib import request
import pandas as pd
import numpy as np
```

```python
data = requests.get('https://www.kpetro.or.kr/lay1/S1T602C603/contents.do')
soup = bs(data.text, 'html.parser')
a = soup.select('#subContent > section > div > table > tbody')

list_a = [a[i].text for i in np.arange(len(a))]
list_a = list_a[0].split( )
```

```python
a = list_a[1:7]
b = list_a[8:14]
c = list_a[15:21]
d = list_a[22:28]
e = list_a[29:35]
f = list_a[36:42]
g = list_a[43:49]
h = list_a[50:56]
i = list_a[57:63]
j = list_a[64:70]
k = list_a[71:77]
l = list_a[78:84]
m = list_a[85:91]
n = list_a[92:98]
o = list_a[99:105]
p = list_a[106:112]
q = list_a[113:119]
r = list_a[120:126]
```

```python
df = pd.DataFrame(columns=['강원도','경기도','경상남도','경상북도','광주광역시','대구광역시','대전광역시','부산광역시','서울특별시','세종특별자치시','울산광역시','인천광역시','전라남도','전라북도','제주도','충청남도','충청북도','계'])
df['강원도'] = a
df['경기도'] = b
df['경상남도'] = c
df['경상북도'] = d
df['광주광역시'] = e
df['대구광역시'] = f
df['대전광역시'] = g
df['부산광역시'] = h
df['서울특별시'] = i
df['세종특별자치시'] = j
df['울산광역시'] = k
df['인천광역시'] = l
df['전라남도'] = m
df['전라북도'] = n
df['제주도'] = o
df['충청남도'] = p
df['충청북도'] = q
df['계'] = r
```

```python
df = df.T
df.columns =[2017,2018,2019,2020,2021,2022]
df
```

```python
df.to_csv('석유판매업소 현황 지역별 2017-2022.9.csv')
```

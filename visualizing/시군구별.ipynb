```python
loc=[]
for i in df["주소"]:
    loc.append(i.split())
```

```python
loc2 = []
for i in loc:
    if "경기" == i[1]:
        loc2.append("경기도")
    elif "영동" == i[1]:
        loc2.append("영동군")
    elif "천안" == i[1]:
        loc2.append("천안시")
    elif "음성" == i[1]:
        loc2.append("음성군")
    elif "광주" == i[1]:
        loc2.append("광주시")
    elif "익산" == i[1]:
        loc2.append("익산시")
    elif "김해" == i[1]:
        loc2.append("김해시")
    elif "칠곡" == i[1]:
        loc2.append("칠곡군")
    elif "밀양" == i[1]:
        loc2.append("밀양시")
    elif "구좌읍" == i[1]:
        loc2.append("제주시")
    elif "남양주" == i[1]:
        loc2.append("남영주시")
    elif "남원" == i[1]:
        loc2.append("남원시")
    else:
        loc2.append(i[1])
```

```python
pd.options.display.max_rows = 100
```

```python
sigun = df[["지역","주소","휘발유","경유"]]
sigun["주소"]=loc2
sigun=sigun.groupby(["지역","주소"])["휘발유","경유"].mean().astype(int)
sigun=pd.DataFrame(sigun)
sigun=sigun.reset_index()
```

```python
sigun_1 = sigun.sort_values(ascending=False, by='휘발유')
sigun_2 = sigun.sort_values(ascending=False, by='경유')
# sigun_1.to_excel('시군구별 휘발유.xlsx')
sigun_2.to_excel('시군구별 경유.xlsx')
```
